```python
#1번
import math
import numpy
X = int(input("첫 번째 숫자를 입력하시오:"))
Y = int(input("두 번째 숫자를 입력하시오:"))
Z = int(input("세 번째 숫자를 입력하시오:"))
NUM = [X, Y, Z]
avg = (X + Y + Z) / 3
std = numpy.std(NUM)
print("%d, %d, %d의 평균은 %f 입니다." %(X, Y, Z, avg))
print("%d, %d, %d의 표준편차는 %f 입니다." %(X, Y, Z, std))
```

```python
#3번
import numpy as np
from matplotlib import pyplot as plt

def f(t):
    return np.cos(10 * np.pi * t) + 0.5 * np.cos(2 * np.pi * t) * np.cos(10 * np.pi * t)
t1 = np.arange(-1, 1, 1 / 10000)
plt.plot(t1, f(t1))
plt.show()
```
